```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data
df = pd.read_csv("personality_dataset.csv")

# Ambil kolom yang digunakan, lalu hapus NaN
df_clean = df[['Friends_circle_size', 'Time_spent_Alone']].dropna()

# Siapkan variabel X dan Y
X = df_clean[['Friends_circle_size']]
y = df_clean['Time_spent_Alone']

# Buat dan latih model regresi linier
model = LinearRegression()
model.fit(X, y)

# Prediksi
y_pred = model.predict(X)

# Visualisasi dengan matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data Asli')
plt.plot(X, y_pred, color='red', linewidth=2, label='Garis Regresi')
plt.title("Regresi Linier: Friends Circle Size vs Post Frequency")
plt.xlabel("Friends Circle Size")
plt.ylabel("Time Spent Alone")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("personality_dataset.csv")

# Ambil dua kolom: Friends_circle_size dan Time_spent_Alone, lalu drop NaN
df_clean = df[['Friends_circle_size', 'Time_spent_Alone']].dropna()

# Ambil nilai sebagai array
x = df_clean['Friends_circle_size'].values
y = df_clean['Time_spent_Alone'].values

# Hitung koefisien regresi secara manual
n = len(x)
mean_x = np.mean(x)
mean_y = np.mean(y)

# Rumus regresi linier: y = mx + c
# m = slope (kemiringan), c = intercept (titik potong sumbu Y)
m = np.sum((x - mean_x) * (y - mean_y)) / np.sum((x - mean_x)**2)
c = mean_y - m * mean_x

# Prediksi nilai y
y_pred = m * x + c

# Visualisasi
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x, y_pred, color='red', label='Garis Regresi')
plt.title("Regresi Linier: Friends Circle Size vs Time Spent Alone")
plt.xlabel("Friends Circle Size")
plt.ylabel("Time Spent Alone")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Cetak nilai m dan c
print(f"Persamaan regresi: y = {m:.2f}x + {c:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("personality_dataset.csv")

# Ubah personality menjadi angka: Introvert = 0, Extrovert = 1
df['Personality_numeric'] = df['Personality'].map({'Introvert': 0, 'Extrovert': 1})

# Hitung perbandingan jumlah masing-masing tipe
count = df['Personality_numeric'].value_counts().sort_index()

# Tampilkan hasil numerik
print("Perbandingan jumlah:")
print(f"Introvert (0): {count[0]}")
print(f"Extrovert (1): {count[1]}")

# Buat bar chart
plt.figure(figsize=(6, 4))
plt.bar(['Introvert (0)', 'Extrovert (1)'], count, color=['blue', 'orange'])
plt.title("Perbandingan Jumlah Introvert vs Extrovert")
plt.ylabel("Jumlah Orang")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("personality_dataset.csv")

# Encode Personality ke angka
df['Personality_numeric'] = df['Personality'].map({'Introvert': 0, 'Extrovert': 1})

# Ambil kolom yang diperlukan dan hapus NaN
df_clean = df[['Friends_circle_size', 'Personality_numeric']].dropna()

# Ambil nilai array
x = df_clean['Friends_circle_size'].values
y = df_clean['Personality_numeric'].values

# Hitung parameter regresi
mean_x = np.mean(x)
mean_y = np.mean(y)
m = np.sum((x - mean_x) * (y - mean_y)) / np.sum((x - mean_x)**2)
c = mean_y - m * mean_x

# Prediksi
y_pred = m * x + c

# Visualisasi
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x, y_pred, color='red', label='Garis Regresi')
plt.title("Regresi Linier: Friends Circle Size vs Personality")
plt.xlabel("Friends Circle Size")
plt.ylabel("Personality (0 = Introvert, 1 = Extrovert)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Tampilkan persamaan garis
print(f"Persamaan regresi linier: y = {m:.4f}x + {c:.4f}")
```
