```python
import pandas as pd
df = pd.read_csv('supermarket_buyer.csv')
df.head()
```

```python
print("Statistik Pengunjung:")
print(df['Visitor'].describe())
print("\nStatistik Pembeli:")
print(df['Buyer'].describe())
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = df[['Visitor']] # Variabel independen harus 2D
y = df['Buyer']       # Variabel dependen

model = LinearRegression()
model.fit(X, y)

beta_0 = model.intercept_
beta_1= model.coef_[0]

print(f"Persamaan Regresi: Pengunjung = {beta_0:.2f} + {beta_1:.2f} * Pembeli")
#Setiap tambahan 1 Pengunjung, meningkatkan peluang pembeli sebesar 0,40%. 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.regplot(x='Visitor', y='Buyer', data=df, ci=None, line_kws={"color": "red"})
plt.title('Hubungan Pengunjung dan Pembeli di Supermarket', fontsize=16)
plt.xlabel('Buyer', fontsize=12)
plt.ylabel('Visitor', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.annotate(f'Pengunjung = {beta_0:.2f} + {beta_1:.2f} * Pembeli',
             xy=(0.55, 0.9), xycoords='axes fraction', fontsize=12, color='blue')
plt.show()
```
