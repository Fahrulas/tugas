```python
import pandas as pd
df = pd.read_csv('city_weather_data.csv')
df.head(10)
```

```python
print("Statistika Suhu:")
print(df['Temperature (°C)'].describe())
print("\nStatistika Kelembapan:")
print(df['Humidity (%)'].describe())
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

x = df[['Temperature (°C)']]
y = df['Humidity (%)']

model = LinearRegression()
model.fit(x, y)

beta_0 = model.intercept_
beta_1 = model.coef_[0]

print(f"Persamaan Regresi : Kelembapan = {beta_0:.2f} + {beta_1:.2f} * Suhu")
#jika suhu bertambah 1c maka kelembapan akan turun sebesar 2,22%
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.regplot(x='Temperature (°C)' , y = 'Humidity (%)' , data=df, ci=None, line_kws={"color": "green"})
plt.title('Hubungan suhu dan kelembapan diberbagai kota', fontsize=16)
plt.xlabel('Suhu (°C)', fontsize=12)
plt.ylabel('Kelembapan (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.annotate(f'Kelembapan = {beta_0:.2f} + {beta_1:.2f} * Suhu',
             xy=(0.55, 0.9), xycoords='axes fraction', fontsize=12,color='blue')
plt.show()
```
